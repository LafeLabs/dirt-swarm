```python
import time
import cv2

# 1. Turn on the camera
cam = cv2.VideoCapture(1)#back camera
#cam = cv2.VideoCapture(0)#front camera

# FIX 1: Give the camera hardware 2 seconds to warm up and adjust its light settings
time.sleep(2)

try:
    if not cam.isOpened():
        print("Error: Could not open the camera.")
    else:
        # 2. Read the frame
        ret, frame = cam.read()

        if ret:
            # Save and show the image
            cv2.imwrite("laptop_photo.jpg", frame)
            print("Success: Image saved!")
            cv2.imshow("Captured Image", frame)
            
            # FIX 2: Wait 3 seconds, then close the window automatically
            cv2.waitKey(3000)
        else:
            print("Error: Could not grab a photo frame.")

finally:
    # FIX 3: This block ALWAYS runs, even if the code crashes. 
    # This guarantees the camera turns off and the light goes out!
    print("Turning off camera now...")
    cam.release()
    cv2.destroyAllWindows()
```
